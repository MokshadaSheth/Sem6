```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
data = pd.read_csv("Iris.csv")

print(data.head())
data = data.dropna()
```

```python
X = data.iloc[:, :-1]   # all columns except last
y = data.iloc[:, -1]    # last column (species)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
print(X_test)
```

```python
model = GaussianNB()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Predicted Values:")
print(y_pred)
```

```python
'''
Class 0 → Setosa

Class 1 → Versicolor

Class 2 → Virginica
'''
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
```

```python
cm = np.array(cm)

n = cm.shape[0]

for i in range(n):

    TP = cm[i][i]
    FP = sum(cm[:, i]) - TP
    FN = sum(cm[i, :]) - TP
    TN = cm.sum() - (TP + FP + FN)

    print("Class", i)
    print("TP:", TP)
    print("FP:", FP)
    print("FN:", FN)
    print("TN:", TN)
    print()
```

```python
#When a confusion matrix is multi-class (3×3), we compute TP, FP, TN, FN for each class using one-vs-all method.

for i in range(n):

    TP = cm[i][i]
    FP = sum(cm[:, i]) - TP
    FN = sum(cm[i, :]) - TP
    TN = cm.sum() - (TP + FP + FN)

    accuracy = (TP + TN) / cm.sum()
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    error_rate = (FP + FN) / cm.sum()

    print("Class", i)
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("Error Rate:", error_rate)
    print()
```

```python
new_samples = pd.DataFrame(
    [
        [1, 5.1,3.5,1.4,0.2],
        [2, 6.0,2.9,4.5,1.5],
        [3, 6.5,3.0,4.8,1.8]
    ],
    columns=X.columns
)

predictions = model.predict(new_samples)

print("Predictions:", predictions)
```
