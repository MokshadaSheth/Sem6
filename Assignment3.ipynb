```python
# Count of each species
species_counts = df["Species"].value_counts()
print(species_counts)
```

```python
def species_statistics(dataframe):
    species_list = dataframe["Species"].unique()
    
    for species in species_list:
        print(f"\nStatistics for {species}:\n")
        
        stats = (
            dataframe[dataframe["Species"] == species]
            .drop(columns=["Species"])
            .describe()
            .transpose()
        )
        
        print(stats)

# Call the function
species_statistics(df)
```
