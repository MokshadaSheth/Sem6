```python
#Replacing Mean for null values. 
#Calculate mean of each column
math_score_mean = int(data['math score'].mean())
print("Math score mean:", math_score_mean)
reading_score_mean = int(data['reading score'].mean())
print("Reading score mean:", reading_score_mean)
writing_score_mean = int(data['writing score'].mean())
print("Writing score mean:", writing_score_mean)
```

```python
sorted_reading_score=sorted(data['reading score'])
sorted_reading_score
```

```python
#Calculating quartiles
q1=np.percentile(sorted_reading_score,25)
q3=np.percentile(sorted_reading_score,75)
print(q1,q3)
```

```python
IQR=q3-q1
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
# Finding outliers using lb and ub of quartile
r_outlier=[]
for i in sorted_reading_score:
     if i<lwr_bound or i >upr_bound:
         r_outlier.append(i)
r_outlier
# outlier_indices = data[(data['reading score'] < lwr_bound) | (data['reading score'] > upr_bound)].index
# print(outlier_indices)
```

```python
#Removing outliers using trimming
for i in r_outlier:
    data['reading score'].drop(i,inplace=True)

#checking is outlier deleted
r_outlier=[]
for i in data['reading score']:
     if i<lwr_bound or i >upr_bound:
         r_outlier.append(i)
if len(r_outlier) == 0:
    print("No outliers remaing")
else:
    print("Still have outliers")
    print(r_outlier)
```
