```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python

# Create box plot
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=df,
    x='Sex',
    y='Age',
    hue='Survived'
)

# Labels and title
plt.title('Age Distribution by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.show()
```

```python
#Inference from the above statistics:
#1. Scatter Plot
plt.figure(figsize=(10, 7)) # Optional: Adjusts plot size

# Create the scatter plot
sns.scatterplot(
    x='Age',
    y='Fare',
    hue='Survived',   # Color points based on survival status (0=No, 1=Yes)
    style='Sex',      # Use different marker shapes for sex (male/female)
    data=df,
    alpha=0.6         # Adds transparency to better see overlapping points
)

# Add titles and labels
plt.title('Age vs. Fare on the Titanic (colored by Survival, styled by Sex)')
plt.xlabel('Age (Years)')
plt.ylabel('Fare ($)')
plt.legend(title='Survived', loc='upper right', labels=['Did Not Survive', 'Survived']) # Customize legend labels
plt.show()
```

```python
'''
Plots:
1. Relational: 
    i.Scatter
    ii.Line
2. Distributional:
    i. KDE
    ii. Histogram
    iii. Joint
    iv. Pair
3. Categorical:
    i. Boxplot
    ii. Barplot
    iii. Valin
    iv. Swarn
'''
```

```python
sns.lineplot(x='Age', y='Fare', data=df)
plt.title("Relational: Line Plot")
plt.show()
```

```python
'''A Kernel Density Estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, 
analagous to a histogram but using continuous density curves '''
sns.kdeplot(x='Age', hue='Survived', fill=True, data=df)
plt.title("Distributional: KDE Plot")
plt.show()
```

```python
sns.histplot(x='Age', bins=20, kde=False, data=df)
plt.title("Distributional: Histogram")
plt.show()
```

```python
'''
A joint plot is a Seaborn visualization in Python that displays the relationship between two numerical variables while 
simultaneously showing their individual (marginal) distributions
Common kind Parameter Options 
scatter: (Default) Displays data points.
kde: Kernel Density Estimation, smooths the distribution.
hex: Hexagonal binning, useful for large datasets.
hist: Histogram representation.
reg: Adds a linear regression line. 
'''
sns.jointplot(x='Age', y='Fare', data=df, kind='reg') # kind='reg' adds a regression line
plt.show()
```

```python
''' A pair plot (or "scatterplot matrix") is a grid of scatter plots used to visualize pairwise relationships 
between multiple numerical variables in a dataset, helping to identify correlations and patterns'''
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.show()
```

```python
#Shows the mean value of each
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Categorical: Barplot")
plt.show()
```

```python
sns.violinplot(x='Pclass', y='Age', hue='Sex', split=True, data=df)
plt.title("Categorical: Violin Plot")
plt.show()
```

```python
# scatter plot where points are adjusted to avoid overlap
sns.swarmplot(x='Pclass', y='Age', data=df, size=3)
plt.title("Categorical: Swarm Plot")
plt.show()
```
